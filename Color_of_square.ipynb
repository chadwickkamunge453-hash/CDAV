```python
x = input("Enter a letter ")
if x=="a" or x=="c" or x=="e" or x=="g":
  letter=1
else:
  letter=2
y = input("Enter a number ")
if y=="1" or y=="3" or y=="5" or y=="7":
  number=1
else:
  number=2
z = (letter+number)

if z==2 or z==4:
  print("black")
else:
  print("white")




```
