```python
word = input("Enter a word ")
left = 0
right = len(word) - 1
palindrome=True
while left<right:
  if word[left]!=word[right]:
    palindrome=False
    break
  left += 1
  right -= 1

if palindrome:
  print("this is a palindrome")
else:
  print("This is not a palindrome")
```
