```python
x = int(input("Enter a noise level in decibels "))
if x>130:
  print("Above jackhammer noise level")
elif x==130:
  print("jackhammer noise level")
elif x>106:
  print("Above gas lawnmower noise level but below jackhammer")
elif x==106:
  print("gas lawnmower noise level")
elif x>70:
  print("Above alarm clock noise level but below gas lawnmower")
elif x==70:
  print("alarm clock noise level")
elif x>40:
  print("Above quiet room noise level but below alarm clock")
elif x==40:
  print("Above quiet room noise level")
else:
  print("below quiet room noise level")
```
