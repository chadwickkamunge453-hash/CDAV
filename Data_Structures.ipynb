Dictionaries

```python
student_grades ={
    "math" : 80,
    "english" : 75,
    "chemistry" : 66,
}
print(type(student_grades))
```

```python
math_results = student_grades["math"]
print(math_results)
```

```python
student_grades["chemistry"] = 69
print(student_grades)
```

```python
del student_grades["chemistry"]
print(student_grades)
```

```python
student_grades.update({"chemistry":68})
print(student_grades)
```

```python
for key in student_grades.keys():
  print(key)
```

```python
for value in student_grades.values():
 print(value)
```

```python
for item in student_grades.items():
  print(item)
```

```python
for key, value  in student_grades.items():
  print(key, "contains", value)
```
