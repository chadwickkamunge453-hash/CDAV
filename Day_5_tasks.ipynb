```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
choice = input("Enter E for encoding or D for decoding: ").upper()
letter = input("Enter a letter: ").lower()
shift = 3
message = ""
while letter != "stop":
    if letter in alphabet:
        index = alphabet.index(letter)
        if choice == "E":
            new_index = (index + shift) % 26
        elif choice == "D":
            new_index = (index - shift) % 26
        else:
            print("Invalid choice!")
            break
        message += alphabet[new_index]
    elif letter == " ":
        message += " "
    else:
        print("Invalid input")
    letter = input("Enter the next letter (or type 'stop' to finish): ").lower()
if choice == "E":
    print("Encrypted message:", message)
elif choice == "D":
    print("Decoded message:", message)
```

```python
word = input("Enter a word ")
left = 0
right = len(word) - 1
palindrome=True
while left<right:
  if word[left]!=word[right]:
    palindrome=False
    break
  left += 1
  right -= 1

if palindrome:
  print("this is a palindrome")
else:
  print("This is not a palindrome")
```

```python
n = int(input("Enter a positive integer(n) "))
m = int(input("Enter a negative integer(m) "))

if n<m:
  d = n
else:
  d = m

while n%d!=0 or m%d!=0:
  d=d-1

print("The GCD of n and m is",d)
```
