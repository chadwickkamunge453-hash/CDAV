```python
x = input("How much did your meal cost? ")
Tip = float(x)*0.18
Tax = float(x)*0.16
Total = float(x)+float(Tax)+float(Tip)
print(f"Tax amount is {Tax:.2f} Shillings")
print(f"Tip is {Tip:.2f} Shillings")
print(f"Total price is {Total:.2f} Shillings")
```
